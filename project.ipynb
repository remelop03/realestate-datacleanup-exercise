#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
import pandas as pd
import matplotlib.pyplot as plt
ds = pd.read_csv('assets/real_estate.csv', sep=';')
south_belt = ds[ds['level5'].isin(["Fuenlabrada","Leganés","Getafe","Alcorcón"])]
south_belt = south_belt.dropna(subset=["price"],axis="index")
south_belt['pps'] = south_belt['price'] / south_belt['surface']
getafe_pps = south_belt[south_belt['level5'] == 'Getafe']['pps'].mean()
alcorcon_pps = south_belt[south_belt['level5'] == 'Alcorcón']['pps'].mean()
print(f"The price per square meter (price/m2) in Getafe is {getafe_pps}")
print(f"The price per square meter (price/m2) in Alcorcón is {alcorcon_pps}")

if getafe_pps > alcorcon_pps:
    print("The price per square meter in Getafe is higher than in Alcorcón.")
elif getafe_pps < alcorcon_pps:
    print("The price per square meter in Getafe is lower than in Alcorcón.")
else:
    print("The price per square meter is the same in Getafe and Alcorcón.")
```
